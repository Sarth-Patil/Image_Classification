```python
import keras
import numpy as np
from keras.applications import vgg16
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.applications.imagenet_utils import decode_predictions
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
model = vgg16.VGG16(weights='imagenet')
```

```python
img= 'panda.jpg'
image = load_img(img, target_size=(224, 224))
#convert the PIL image (width, height) to a NumPy array (height, width, channel)
np_image = img_to_array(image)
```

```python
#Convert the image into 4D Tensor (samples, height, width, channels) 
#adding an extra dimension to the axis 0.
ip_image = np.expand_dims(np_image, axis=0)
```

```python
plt.imshow(np.uint8(ip_image[0]))
```

```python
processed_image = vgg16.preprocess_input(ip_image.copy())
```

```python
predictions = model.predict(processed_image)
label = decode_predictions(predictions)
print("label = ", label)
```

```python
result = str(label[0][0][1])
print("The image is of",result)
```
